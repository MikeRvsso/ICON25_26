# Introduzione

Il software di supporto ha come obiettivo di sviluppare un Assistente Intelligente per la Logistica che aiuti a ottimizzare l'intero processo di consegna giornaliero. Il sistema agisce come un coordinatore capace di analizzare il contesto operativo per prevedere i tempi di percorrenza.  organizzare intelligentemente il carico del veicolo per massimizzare le consegne e calcola il percorso ottimale su mappa. Infine, attraverso un motore probabilistico, valuta i rischi di mancata consegna, offendo un supporto decisionale completo.
